```python
import pandas as pd
from unidecode import unidecode

# Leer el archivo CSV
df = pd.read_csv("Mazda_manual.csv")

# Crear un conjunto para almacenar modelos únicos
modelos_unicos = set()
make = []
modelos_enlaces = {}

# Iterar sobre cada fila del DataFrame
for index, row in df.iterrows():
    if not make:
        make = row["Make"]
        make = unidecode(make.upper())

    # Obtener el modelo de la fila actual y limpiarlo
    modelo = row["Model"]
    modelo = unidecode(modelo.upper())  # Convertir a minúsculas y eliminar caracteres acentuados
    modelo = ''.join(e for e in modelo if e.isalnum())  # Eliminar caracteres no alfanuméricos
    # Agregar el modelo al conjunto de modelos únicos
    modelos_unicos.add(modelo)

    if modelo not in modelos_enlaces:
        # Si el modelo es nuevo, almacenar el enlace correspondiente en el diccionario
        modelos_enlaces[modelo] = row["URL"]

# Convertir el conjunto a una lista y ordenarla alfabéticamente
modelos_unicos_ordenados = sorted(list(modelos_unicos))

# Imprimir los modelos únicos ordenados alfabéticamente
for modelo in modelos_unicos_ordenados:
    print(modelo)
    print(modelos_enlaces[modelo])

# Crear un DataFrame de Pandas con los modelos únicos
df_modelos = pd.DataFrame(modelos_unicos_ordenados, columns=["Modelo"])

nombre_archivo = f"{make}_MODELS.csv"

# Guardar el DataFrame en un archivo CSV
# df_modelos.to_csv(nombre_archivo, index=False)  # El parámetro index=False evita que se agregue una columna de índice al archivo CSV
```
