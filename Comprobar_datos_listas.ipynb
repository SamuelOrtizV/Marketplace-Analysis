```python
import pandas as pd

# Cargamos los datos
# Leer el archivo CSV

nombre_archivo = "KIA _MANUAL.csv"

df = pd.read_csv(nombre_archivo)

# Mostrar las filas de nombre, precio y url

urls = df['URL']
precio = df['Price']
nombre = df['Titel']

# Mostrar las filas de nombre, precio y url

for i in range(len(urls)):
    print(f"Nombre: {nombre[i]}")
    print(f"Precio: {precio[i]}")
    print(f"URL: {urls[i]}")
    print("")
```
